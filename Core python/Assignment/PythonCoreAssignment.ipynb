### Write a program to enter P, T, R and calculate Compound Interest.


##### Algorithm

```python
# Start
# Input Principal Amount (P)
# Input Time (T)
# Input Rate of Interest (R)
# Calculate Amount (A) using:

# A = P * (1 + R / 100) ** T

# Calculate Compound Interest:

# CI=A−P

# Display Amount and Compound Interest
# End
```

```python
P = float(input('Enter Principle Amount '))
T = float(input('Enter Time '))
R = float(input('Enter Rate of Interest '))

A = P * ( 1 + R / 100) ** T 

CI = A - P

print(f'Compound Interest = {CI}')
```

### Program to Find the Roots of a Quadratic Equation



##### Algorithm


```python
# Start
# Input values of a, b, and c
# Calculate the discriminant:
# d = b² - 4ac
# Calculate the roots:
# root1 = (-b + √d) / (2a)
# root2 = (-b - √d) / (2a)
# Display the roots
# Stop
```

```python
import math

a = float(input("Enter value of a: "))
b = float(input("Enter value of b: "))
c = float(input("Enter value of c: "))

d = b**2 - 4*a*c

root1 = (-b + math.sqrt(d)) / (2*a)
root2 = (-b - math.sqrt(d)) / (2*a)

print(f"Root 1 = {root1:.2f}")
print(f"Root 2 = {root2:.2f}")
```

### WAP to calculate total salary of employee based on basic, da=10% of basic, ta=12% of basic, hra=15% of basic.

```python
# DA = 10% of Basic Salary
# TA = 12% of Basic Salary
# HRA = 15% of Basic Salary
# Steps:
# Start
# Input Basic Salary (basic)

# Calculate DA:

# DA = basic × 10 / 100

# Calculate TA:

# TA = basic × 12 / 100

# Calculate HRA:

# HRA = basic × 15 / 100

# Calculate Total Salary:

# Total Salary = basic + DA + TA + HRA

# Display DA, TA, HRA, and Total Salary
# Stop
```

### Find the sum of three-digit number.

##### Algorithm

```python
# Start
# Input a three-digit number (num)

# Find the unit digit:

# unit = num % 10

# Find the tens digit:

# tens = (num // 10) % 10

# Find the hundreds digit:

# hundreds = num // 100

# Calculate the sum:

# sum = hundreds + tens + unit

# Display the sum
# Stop
```

```python
# Input a three-digit number
num = int(input("Enter a three-digit number: "))

# Extract digits
unit = num % 10
tens = (num // 10) % 10
hundreds = num // 100

# Calculate sum of digits
sum_digits = hundreds + tens + unit

# Display result
print("Sum of digits =", sum_digits)
```

### Write a program to reverse three-digit number.

##### Algorithm

```python
# Start
# Input a three-digit number (num)

# Extract the unit digit:

# unit = num % 10

# Extract the tens digit:

# tens = (num // 10) % 10

# Extract the hundreds digit:

# hundreds = num // 100

# Form the reversed number:

# reverse = (unit * 100) + (tens * 10) + hundreds

# Display the reversed number
# Stop
```

```python
# Input a three-digit number
num = int(input("Enter a three-digit number: "))

# Extract digits
unit = num % 10
tens = (num // 10) % 10
hundreds = num // 100

# Reverse the number
reverse = (unit * 100) + (tens * 10) + hundreds

# Display result
print("Reversed number =", reverse)
```
